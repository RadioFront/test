```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Correlate action and observation
idx=0
data_=np.genfromtxt('px4_rl_observation1200_1200.txt', delimiter='\t')
#print(data_)
data=data_[:,idx]
#print(data)
max=data.max()
min=data.min()
#print(max)
#print(min)
data1=(data-((max+min)/2)) / (max-min) *2

idx=11
data_=np.genfromtxt('px4_rl_action1200_1200.txt', delimiter='\t')
data=data_[:,idx]
max=data.max()
min=data.min()
data2=(data-((max+min)/2)) / (max-min) *2

i=range(np.size(data,0))
t=np.array(i)/10
f=plt.figure()
f.set_figwidth(10)
f.set_figheight(2)
plt.title("Gyroscope X-axis acceleration and XY velocity controller P coefficient at iteration 1400")
plt.xlabel("Time (sec)")
plt.ylabel("Scaled output")
plt.plot(t,data1)
plt.plot(t,data2)
plt.show
```

```python
#Evaluation results over time
max_idx=1550
f=plt.figure()
f.set_figwidth(10)
f.set_figheight(2)
plt.title("Evaluation over iteration")
plt.xlabel("Iteration")
plt.ylabel("Average distance")

t = range(0, max_idx,50)
mean=[]
max=[]
std=[]
for i in range(0, max_idx, 50):
    data_=np.genfromtxt('px4_rl_output1200_%d.txt' % i, delimiter='\t')
    mean.append(data_[:,0].mean())
    std.append(data_[:,0].std())
    max.append(data_[:,1].max())

plt.plot(t,mean)
plt.plot(t,np.array(mean) + np.array(std))
plt.plot(t,np.array(mean) - np.array(std))
#plt.plot(t,max)
plt.show

print(data_[:,0].mean())
print(data_[:,0].std())
print(data_[:,0].std()/data_[:,0].mean())
```
